Изучив уникальные значения, можем заметить, что часть дублей вызвана различным написанием буквы "ё". Приведем к единообразию - везде оставим написание через "е".

Также можем заметить, что имеются дубли, вызванные различным наименованием типов населенных пунктов.

Чтобы получить список повторяющихся значений, разобьем уникальные наименования на отдельные слова и посчитаем их частотность. 

```python
unique_locality_name_list = []
for name in unique_locality_name:
    for word in name.split():
        unique_locality_name_list.append(word)
unique_locality_name_list    
```

```python
print(Counter(unique_locality_name_list)) 
```

Составим список слов от которых стоит избавится. Это слова, которые не являются именами собственными и часто повторяются.

```python
del_list = ['поселок', 'деревня', 'городского', 'типа', 'городской', 'село', 'станции', 'товарищество', 'садовое', 'коттеджный']
```

Напишем функцию, которая удаляет из наименований населенных пунктов слова из списка нежелательных.

```python
def del_func(name):
    name_list = []
    for word in name.split():
        if word not in del_list:
            name_list.append(word)
    return ' '.join(name_list)
            
```

Проверим работу функции.

```python
x = 'деревня Новое Девяткино'
```

```python
print(del_func(x))
```

Функция работает корректно, применим её ко всему объему данных.

```python
data['locality_name'] = data['locality_name'].apply(del_func)
```

Проверим результат.

```python
data['locality_name'].value_counts().head(30)
```

```python
unique_locality_name = data['locality_name'].unique()
unique_locality_name
```

```python
len(unique_locality_name)
```

**этаж квартиры, варианты — первый, последний, другой**

Напишем функцию, определяющую категорию по этажу расположения квартиры.

```python
def floor_group(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return 'first'
    if floor == floors_total:
        return 'last'
    return 'other'
```

Проверим работу функции.

```python
r = data.loc[0]
print(floor_group(r))
```

Функция работает корректно, применим её ко всему объему данных.

```python
data['floor_group'] = data.apply(floor_group, axis=1)
```

Проверим результат.

```python
data['floor_group'].value_counts()
```

Результат корректный.
